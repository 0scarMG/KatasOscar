Ejercicio 2

```python
# Lista de planetas
planets = ['Mercurio', 'Venus', 'Tierra', 'Marte', 'Jupiter', 'Saturno', 'Urano', 'Neptuno']
```

```python
# Solicitamos el nombre de un planeta
introducir_planeta = input('Introduce el nombre de un planeta, Por favor iniciar con mayuscula')
```

```python
# Busca el planeta en la lista
planet_index = planets.index(introducir_planeta)
print(planet_index)
```

```python
# Muestra los planetas m√°s cercanos al sol
print('Aqui estan los planetas mas cerca al sol que: '+ introducir_planeta)
print(planets[0:planet_index])
```

```python
print('Aqui estan los planetas mas lejanos al sol que: '+ introducir_planeta)
print(planets[planet_index + 1:])
```

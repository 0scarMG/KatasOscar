```python
# Función para leer 3 tanques de combustible y muestre el promedio

def reporte(tanque1, tanque2, tanque3):
    total_porcentaje = (tanque1 + tanque2 + tanque3) / 3
    return f"""Reporte de combustible:
    Total: {total_porcentaje}%
    Tanque principal: {tanque1}%
    Tanque externo: {tanque2}%
    Tanque de Hidrogeno: {tanque3}% 
    """
```

```python
# Llamamos a la función que genera el reporte print(funcion(tanque1, tanque2, tanque3))

print(reporte(80, 70, 85))
```

```python
# Función promedio 
def average(values):
    total = sum(values)
    number_of_items = len(values)
    return total / number_of_items

average([80, 85, 81]) 
```

```python
# Actualiza la función
def reporte(tanque1, tanque2, tanque3):
    return f"""Reporte de combustible:
    Total : {average([tanque1, tanque2, tanque3])}%
    Tanque principal: {tanque1}%
    Tanque externo: {tanque2}%
    Tanque de hidrogeno: {tanque3}% 
    """


print(reporte(88, 76, 70))
```

```python
# Escribe tu nueva función
def reporte_mision(destino, *minutos, **reserva_conbustible):
    reporte = f"""
    mision con destino a {destino}
    total de tiempo de viaje: {sum(minutos)} minutos
    total de combustible: {sum(reserva_conbustible.values())}
    """
    for tanque, gallones in reserva_conbustible.items():
        reporte += f"{tanque} tanque --> {gallones} gallones restantes\n"
    return reporte
print(reporte_mision("luna", 8, 11, 55, principal=300000, externo=200000))
```

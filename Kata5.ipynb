Ejercicio 2

```python
# Almacenar las entradas del usuario
Planeta1 = input('Introduce la distancia del primer planeta con respecto al sol en km: ')
Planeta2 = input('Introduce la distancia del segundo planeta con respecto al sol en km: ')
```

```python
# Convierte las cadenas de ambos planetas a números enteros
Planeta1 = int(Planeta1)
Planeta2 = int(Planeta2)
```

```python
# Realizar el cálculo y determinar el valor absoluto
distancia_km = Planeta2 - Planeta1
print(distancia_km, 'km')
# Convertir de KM a Millas
distancia_mi = distancia_km * 0.624
print(abs(distancia_mi), 'mi')
```
